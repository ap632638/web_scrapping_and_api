# Extracting data for a time period 
for this we use the url format as /history and parameters start_at & end_at dates.
Here to reduce dataset we add filters like symbols.

```python
time_period = base_url + "/history" + "?start_at=2012-03-03" + "&end_at=2015-03-08" + "&symbols=INR"
time_period
```

```python
response = requests.get(time_period)
response.status_code
```

```python
data = response.json()
print(json.dumps(data, indent = 4))
```

# above the rates are not arranged, to do this we add 
sort_keys=True in dumps method

```python
print(json.dumps(data, indent = 4, sort_keys = True))
```

# Program for a currency convertor

```python
date = input("Please enter a date in format(yyyy-mon-dd) or type 'latest': ")
base = input("Convert from (currency): ")
curr = input("Convert to (currency): ")
quant = float(input("How many {} you want to convert: ".format(base)))

base_url = "https://api.exchangeratesapi.io"
url = base_url + "/" + date + "?base=" + base +"&symbols=" +curr

response = requests.get(url)

if(response.ok is False):
    print("\nError {}:".format(response.status_code))
    print(response.json()['error'])
else:
    data = response.json()
    rate = data['rates'][curr]
    result = quant*rate
    print("\n{0} {1} is equal to {2} {3} based upon exchange rates on {4}.".format(quant,base,result,curr,data['date']))
```
